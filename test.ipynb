```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
samples = np.array([0, 1, 10, 20, 30, 40, 50, 100, 1000])
v1 = np.array([997, 1035, 3587, 7087, 10544, 14004, 18630, 34625, 340678]) * 4
v2 = np.array([1023, 614, 5855, 11659, 18758, 26534, 29034, 57263, 572892])
v3 = np.array([919, 519, 4820, 10193, 14841, 19496, 24433, 48561, 491348])

std = np.std(np.array([5081, 5178, 5058, 5045, 5098, 5122, 5552, 5186, 5149, 5068, 5095, 5093, 5125, 5145, 5095, 5109, 5146]))
sigma = np.ones_like(samples) * std
print(std)
```

```python
plt.figure()
plt.plot(samples, v1, ls='dotted', color='black')
plt.errorbar(samples, v1, yerr=sigma, capsize=2, barsabove=True, ls='None', marker='x', label='V1')

plt.plot(samples, v2, ls='dotted', color='black')
plt.errorbar(samples, v2, yerr=sigma, capsize=2, barsabove=True, ls='None', marker='s', label='V2 (Multi Thread & structs opti)')


plt.plot(samples, v3, ls='dotted', color='black')
plt.errorbar(samples, v3, yerr=sigma, capsize=2, barsabove=True, ls='None', marker='o', label='V3 (doubles to floats)')

plt.xscale('log')
plt.yscale('log')
plt.grid(True, 'both', alpha=.2)
plt.xlabel("Samples")
plt.ylabel("Render time (ms)")
plt.legend()
plt.show()
```
